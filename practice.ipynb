```python
from sklearn import linear_model
import matplotlib.pyplot as plt

feature = [[13],[16] , [20] ,[25]]
labels = [36 , 40 , 44 , 50]


plt.plot(feature , labels , color = "g")
plt.show()
```

```python
lr = linear_model.LinearRegression()
lr = lr.fit(feature , labels)
result = lr.predict([[30] , [10]])
print(result)
```

```python
plt.plot([[30] , [10]] , result ,'r')
plt.xlabel("Age values")
plt.ylabel("weight values")
plt.show()
```
