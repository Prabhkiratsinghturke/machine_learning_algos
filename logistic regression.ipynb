```python
import pandas as pd
import numpy as np
df=pd.read_csv("spam.csv",encoding="latin-1")
df
```

```python
x=df.iloc[:,1].values #independent variables
y=df.iloc[:,0].values  #dependent variables
print(x)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
lb_enco=LabelEncoder()
x=lb_enco.fit_transform(x)
y=lb_enco.fit_transform(y)
print(x)
print(y)
#x=x.reshape(-1,1)
print(x)
```

```python
from sklearn.cross_validation import train_test_split

x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 1/4,random_state = 0)

print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
LRClassifier = LogisticRegression(random_state = 0)
LRClassifier.fit(x_train,y_train)
a = LRClassifier.score(x_train , y_train)
prediction = LRClassifier.predict(x_test)
print(prediction)
a
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
confusionMatrix = confusion_matrix (y_test, prediction)
print(confusionMatrix)
print("Accuracy is:-")
print(accuracy_score(prediction,y_test)*100)
```
